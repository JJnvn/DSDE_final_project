```python
import re
def clean_comment(text):
    # Remove English letters, numbers, and symbols
    text = re.sub(r'[a-zA-Z0-9]', '', text)
    # Remove emojis and special symbols
    text = re.sub(r'[^\u0E00-\u0E7F\s]', '', text)
    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
def prepare_target(df):
    df = df.copy()
    df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed')
    df['last_activity'] = pd.to_datetime(df['last_activity'], format='mixed')
    df['eta_hours'] = (df['last_activity'] - df['timestamp']).dt.total_seconds() / 3600

    # Remove negative durations
    df = df[df['eta_hours'] >= 0]

    # Apply IQR-based outlier removal (3*IQR is more aggressive than standard 1.5*IQR)
    Q1 = df['eta_hours'].quantile(0.25)
    Q3 = df['eta_hours'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    df = df[(df['eta_hours'] >= lower_bound) & (df['eta_hours'] <= upper_bound)].copy()

    # Log-transform the target
    df['log_eta'] = np.log1p(df['eta_hours'])

    return df.reset_index(drop=True)

def clean_and_prepare(df):
    df = df.drop(columns=['ticket_id', 'photo', 'photo_after', 'address'], errors='ignore') 
    df = df.dropna(subset=['organization', 'district','comment']).copy()
    df['type'] = df['type'].fillna('Unknown')
    # df['star'] = df['star'].fillna(0)
    df = df[df['state'] == 'เสร็จสิ้น'].copy()

    # Keep only valid Bangkok districts
    bangkok_districts = [
        'พระนคร', 'ดุสิต', 'หนองจอก', 'บางรัก', 'บางเขน', 'บางกะปิ',
        'ปทุมวัน', 'ป้อมปราบศัตรูพ่าย', 'พระโขนง', 'มีนบุรี', 'ลาดกระบัง',
        'ยานนาวา', 'สัมพันธวงศ์', 'พญาไท', 'ธนบุรี', 'บางกอกใหญ่',
        'ห้วยขวาง', 'คลองสาน', 'ตลิ่งชัน', 'บางกอกน้อย', 'บางขุนเทียน',
        'ภาษีเจริญ', 'หนองแขม', 'ราษฎร์บูรณะ', 'บางพลัด', 'ดินแดง',
        'บึงกุ่ม', 'สาทร', 'บางซื่อ', 'จตุจักร', 'บางคอแหลม',
        'ประเวศ', 'วัฒนา', 'สวนหลวง', 'จอมทอง', 'ดอนเมือง',
        'ราชเทวี', 'ลาดพร้าว', 'วังทองหลาง', 'บางแค', 'หลักสี่',
        'สายไหม', 'คันนายาว', 'สะพานสูง', 'คลองสามวา', 'บางนา',
        'ทวีวัฒนา', 'ทุ่งครุ', 'บางบอน'
    ]
    df = df[df['district'].isin(bangkok_districts)].copy()
    # Clean Thai comment text
    df['comment'] = df['comment'].astype(str).apply(clean_comment)

    df = df.drop(columns=['state', 'last_activity', 'province','timestamp','star' ,'coords','subdistrict'])

    return df.reset_index(drop=True)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import MultiLabelBinarizer
from collections import Counter



# For "organization" (multi-label)
class OrgPipeline(BaseEstimator, TransformerMixin):
    def __init__(self, min_freq=1000):
        self.min_freq = min_freq
        self.common_orgs = None
        self.binarizer = MultiLabelBinarizer(sparse_output=True)

    def fit(self, X, y=None):
        org_lists = X.iloc[:, 0].fillna('').apply(lambda s: [o.strip() for o in s.split(',') if o.strip()])
        flat_orgs = [org for sublist in org_lists for org in sublist]
        counts = Counter(flat_orgs)
        self.common_orgs = {org for org, c in counts.items() if c >= self.min_freq}
        filtered = org_lists.apply(lambda lst: [org for org in lst if org in self.common_orgs])
        self.binarizer.fit(filtered)
        return self

    def transform(self, X):
        org_lists = X.iloc[:, 0].fillna('').apply(lambda s: [o.strip() for o in s.split(',') if o.strip()])
        filtered = org_lists.apply(lambda lst: [org for org in lst if org in self.common_orgs])
        return self.binarizer.transform(filtered)

# For "type" (multi-label)
class TypePipeline(BaseEstimator, TransformerMixin):
    def __init__(self, min_freq=1000):
        self.min_freq = min_freq
        self.common_types = None
        self.binarizer = MultiLabelBinarizer(sparse_output=True)

    def fit(self, X, y=None):
        type_lists = X.iloc[:, 0].fillna('').apply(
            lambda s: [t.strip() for t in s.strip('{}').split(',') if t.strip()]
        )
        flat_types = [t for sublist in type_lists for t in sublist]
        counts = Counter(flat_types)
        self.common_types = {t for t, c in counts.items() if c >= self.min_freq}
        filtered = type_lists.apply(lambda lst: [t for t in lst if t in self.common_types])
        self.binarizer.fit(filtered)
        return self

    def transform(self, X):
        type_lists = X.iloc[:, 0].fillna('').apply(
            lambda s: [t.strip() for t in s.strip('{}').split(',') if t.strip()]
        )
        filtered = type_lists.apply(lambda lst: [t for t in lst if t in self.common_types])
        return self.binarizer.transform(filtered)
```

```python
from pythainlp.tokenize import word_tokenize

def thai_tokenizer(text):
    return word_tokenize(text, keep_whitespace=False)
```

```python
from pythainlp.corpus.common import thai_stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
thai_stopword_list = list(thai_stopwords())  

comment_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        tokenizer=thai_tokenizer,
        stop_words=thai_stopword_list,  
        max_features=300,
        ngram_range=(1, 2),
        min_df=5
    ))
])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import FunctionTransformer
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler


# Build column transformer
preprocessor = ColumnTransformer([
    ('district', OneHotEncoder(sparse_output=True, handle_unknown='ignore'), ['district']),
    ('organization', OrgPipeline(), ['organization']),
    ('type', TypePipeline(), ['type']),
    ('district_avg_eta', StandardScaler(), ['district_avg_eta']), 
    ('organization_avg_eta', StandardScaler(), ['organization_avg_eta']),
    ('count_reopen', StandardScaler(), ['count_reopen']),
    ('comment', comment_pipeline, 'comment')
])
```

```python
# train_df.to_csv("raw.csv", index=False)
train_df = prepare_target(train_df)
train_df = clean_and_prepare(train_df)

scrape_df['comment'] = ''
scrape_df['count_reopen'] = 0
scrape_df = scrape_df[train_df.columns]
train_df = pd.concat([scrape_df, train_df], ignore_index=True)


X = train_df[[ 'district','organization', 'type' , 'count_reopen', 'comment']]
y = train_df['log_eta']
# X.to_csv("cleanv3.csv", index=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Step 1: Create a DataFrame that includes district + log_eta
train_stats = X_train.copy()
train_stats['log_eta'] = y_train

# Step 2: Group by district and compute the average log_eta
avg_eta_by_district = train_stats.groupby('district')['log_eta'].mean()

# Step 3: Map averages to both X_train and X_test
X_train['district_avg_eta'] = X_train['district'].map(avg_eta_by_district)

# Step 4: For districts in test that weren't in train, fill with overall mean
global_avg = y_train.mean()
X_test['district_avg_eta'] = X_test['district'].map(avg_eta_by_district)
X_test['district_avg_eta'] = X_test['district_avg_eta'].fillna(global_avg)
```

```python
# Step 1: Prepare training org data with log_eta
train_org = X_train.copy()
train_org['log_eta'] = y_train

# Step 2: Split organization strings into list
train_org['org_list'] = train_org['organization'].apply(
    lambda s: [org.strip() for org in s.split(',') if org.strip()]
)

# Step 3: Explode into one organization per row
org_exploded = train_org.explode('org_list')

# Step 4: Compute average log_eta per organization
avg_eta_by_org = org_exploded.groupby('org_list')['log_eta'].mean().to_dict()

# Step 5: Apply average to X_train and X_test
def compute_org_avg(org_str, avg_map, default):
    orgs = [o.strip() for o in org_str.split(',') if o.strip()]
    values = [avg_map[o] for o in orgs if o in avg_map]
    return np.mean(values) if values else default

global_avg = y_train.mean()
X_train['organization_avg_eta'] = X_train['organization'].apply(
    lambda s: compute_org_avg(s, avg_eta_by_org, global_avg)
)
X_test['organization_avg_eta'] = X_test['organization'].apply(
    lambda s: compute_org_avg(s, avg_eta_by_org, global_avg)
)
```

```python

X_test.to_csv('x_test.csv')
y_test.to_csv('y_test.csv')
```

```python
from sklearn.pipeline import Pipeline
model_pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('xgb', XGBRegressor(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        verbosity=2
    ))
])

model_pipeline.fit(X_train, y_train)

y_pred = model_pipeline.predict(X_test)

# -------------------------------
# Metrics
# -------------------------------

rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")

# ✅ RMSE = 1.7551
# ⚠️ R² = 0.22 

# add avg district
# ✅ RMSE = 1.7388
# ⚠️ R² = 0.236 

# add avg district, better outlier 3IQR
# ✅ RMSE = 1.7573
# ⚠️ R² = 0.2430 

# add avg org better outlier 3IQR
# ✅ RMSE: 1.6936
# ⚠️ R²: 0.2968

# add reopen
# ✅ RMSE: 1.6646
# ⚠️ R²: 0.3207

# use XGB
# ✅ RMSE: 1.5401
# ⚠️ R²: 0.4169

# use only 1000commons type + webscraping data
# ✅ RMSE: 1.5303
# ⚠️ R²:  0.4211
```
